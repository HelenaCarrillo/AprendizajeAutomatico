### Regresión Linear

Para esta tarea se hará una regresión lineal simple con datos que corresponden al precio de venta de una casa (variable a predecir *y*) y los impuestos a pagar (variable predictora *x*).

Primero se importan las librerías que se necesitarán:

```python
#importar librerias
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

Posteriormente se leen los datos. Los datos se pueden leer de diferentes fuentes, ya sea que se lean directo de un archivo o bien pueden definirse en el mismo código:

```python
#leer datos
#data = pd.read_csv("C:/Users/patty/Documents/Maestria Base de Datos/Aprendizaje_Automatico_priv/Regresion_Tarea1_AA.csv")
```

```python
#definir vectores
x=[4.9176,5.0208,4.5429,4.5573,5.0597,3.891,5.898,5.6039,5.8282,5.3003,6.2712,5.9592,5.05,8.2464,6.6969,7.7841,9.0384,5.9894,7.5422,8.7951,6.0831,8.3607,8.14,9.1416]
y=[25.9,29.5,27.9,25.9,29.9,29.9,30.9,28.9,35.9,31.5,31,30.9,30,36.9,41.9,40.5,43.9,37.5,37.9,44.5,37.9,38.9,36.9,45.8]
```

Una vez que se han definido los datos podemos graficarlos en una matriz de disperción para observar su comportamiento:

```python
#Graficar datos
plt.scatter(x,y)
plt.title("Precio venta de la casa vs Impuestos a pagar")
plt.xlabel("Impuestos a pagar (mil)")
plt.ylabel("Precio venta de la casa (mil)")
plt.show()
```

Se realiza la regresión lineal:

```python
#regresión lineal
# Crear objeto de Regresión Linear
regr = linear_model.LinearRegression()

# Entrenando el modelo
xx=np.array(x)
xxx = xx.reshape(-1,1)
regr.fit(xxx,y)
y_pred=regr.predict(xxx)
```

Podemos imprimir el coeficiente y termino independiente para saber la ecuación resultante así como la media cuadrada del error y la varianza de R2 para ver que tan bien se ajusta el modelo:

```python
# Coeficiente
print('Coeficiente: \n', regr.coef_)
# Valor independiente (b)
print('Termino independiente: \n', regr.intercept_)
#Error Cuadrado Medio
print("Media cuadrada del error: %.2f" % mean_squared_error(y, y_pred))
# Puntaje de Varianza.
print('Varianza R2: %.2f' % r2_score(y, y_pred))
```

De lo anterior se sabe que la ecuación del modelo obtenido es *y*= 3.32*x* + 13.32 y el modelo tiene un ajuste aproximado del 77% el cuál es un ajuste bueno.

También podmeos graficar la linea del modelo obtenido junto con nuestros datos para hacer una comparación visual:

```python
#Imprimir la linea
plt.scatter(x,y)
plt.title("Precio venta de la casa vs Impuestos a pagar")
plt.xlabel("Impuestos a pagar (mil)")
plt.ylabel("Precio venta de la casa (mil)")
plt.plot(x,y_pred, color = "red", label = "y*=3.32x+13.32")
plt.show()
```
