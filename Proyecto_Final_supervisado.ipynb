```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, StackingRegressor
from sklearn.metrics import mean_squared_error as RMSE
from sklearn.metrics import r2_score as r2
from sklearn.metrics import mean_absolute_error as MAE
from sklearn.linear_model import LinearRegression, BayesianRidge, PoissonRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn import linear_model
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(data3)
datascaled = pd.DataFrame(scaled, columns = data3.columns)
```

```python
from timeit import default_timer as timer
from tqdm import tqdm
```

```python
def fit(x, y, reg_model = 'RF', n_components = 3, n_jobs = None):
    match reg_model:
        case 'BayesRidge':
            model = BayesianRidge()
        case 'Boosting':
            model = GradientBoostingRegressor()
        case 'LinearReg':
            model = LinearRegression(n_jobs = n_jobs)
        case 'RF':
            model = RandomForestRegressor(n_jobs = n_jobs)
        case 'KNN':
            model = KNeighborsRegressor(algorithm = 'auto')
        case 'LS':
            model = linear_model.Lasso(alpha=0.1)
        case 'StackingRegressor':
            #https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.StackingRegressor.html#sklearn.ensemble.StackingRegressor
            gb  = GradientBoostingRegressor()
            rf  = RandomForestRegressor(n_jobs = n_jobs)
            gb.fit(x, y)
            rf.fit(x, y)
            model = StackingRegressor([("gb",gb), ("rf",rf)], n_jobs = n_jobs)
    model.fit(x, y)
    return model
```

```python
tsize = [0.7,0.8,0.6]
```

```python
models_ed = ['RF', 'LinearReg', 'BayesRidge', 'Boosting', 'KNN', 'StackingRegressor']
n_jobs = multiprocessing.cpu_count() // 2
results_ed = {
        'model' : [],
        'value' : [],
        'r2':[],
        'time' : []
    }

for j in tsize:
  for i in tqdm(range(10)):
      df_train = datascaled.sample(frac = j)
      x_train = df_train[['Year', 'emisiones_mean', 'diferencia']]
      y_train = df_train[['EMISIONES']]

      df_test = datascaled.drop(df_train.index)
      x_test = df_test[['Year','emisiones_mean','diferencia']]
      y_test = df_test[['EMISIONES']]
      for m in models_ed:
          start = timer()
          model = fit(x_train, y_train.values.ravel(), m, n_jobs = n_jobs)
          end = timer()
          y_pred = model.predict(x_test)
          results_ed['model'].append(m)
          results_ed['value'].append(RMSE(y_test, y_pred, squared = False))
          results_ed['r2'].append(r2(y_test, y_pred))
          results_ed['time'].append(end - start)


df_results_ed = pd.DataFrame(results_ed)
df_results_ed = df_results_ed.sort_values('value')
```

```python
models_ed = ['RF', 'LinearReg', 'BayesRidge', 'Boosting', 'StackingRegressor', 'KNN']
n_jobs = multiprocessing.cpu_count() // 2
results_ed = {
        'model' : [],
        'RMSE' : [],
        'MAE':[],
        'time' : []
    }

for j in tqdm(tsize):
  tscv = TimeSeriesSplit(n_splits=5, test_size=j)

  for g, (train_index, test_index) in enumerate(tscv.split(datascaled)):
      df_train =datascaled.iloc[train_index]
      x_train = df_train[['Year', 'emisiones_mean', 'diferencia']]
      y_train = df_train[['EMISIONES']]

      df_test =datascaled.iloc[test_index]
      x_test = df_test[['Year','emisiones_mean','diferencia']]
      y_test = df_test[['EMISIONES']]

      for m in models_ed:
          start = timer()
          model = fit(x_train, y_train.values.ravel(), m, n_jobs = n_jobs)
          end = timer()
          y_pred = model.predict(x_test)
          results_ed['model'].append(m)
          results_ed['RMSE'].append(RMSE(y_test, y_pred, squared = False))
          results_ed['MAE'].append(MAE(y_pred, y_test))
          results_ed['time'].append(end - start)


df_results_ed = pd.DataFrame(results_ed)
df_results_ed = df_results_ed.sort_values('RMSE')
```
