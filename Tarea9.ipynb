Analizando diversos modelos a la vez

```python
from timeit import default_timer as timer
from tqdm import tqdm
```

```python
def fit(x, y, reg_model = 'RF', n_components = 3, n_jobs = None):
    match reg_model:
        case 'BayesRidge':
            model = BayesianRidge()
        case 'Boosting':
            model = GradientBoostingRegressor()
        case 'LinearReg':
            model = LinearRegression(n_jobs = n_jobs)
        case 'RF':
            model = RandomForestRegressor(n_jobs = n_jobs)
        case 'KNN':
            model = KNeighborsRegressor(algorithm = 'auto')
        case 'StackingRegressor':
            #https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.StackingRegressor.html#sklearn.ensemble.StackingRegressor
            gb  = GradientBoostingRegressor()
            rf  = RandomForestRegressor(n_jobs = n_jobs)
            gb.fit(x, y)
            rf.fit(x, y)
            model = StackingRegressor([("gb",gb), ("rf",rf)], n_jobs = n_jobs)
    model.fit(x, y)
    return model
```

```python
models_ed = ['RF', 'LinearReg', 'BayesRidge', 'Boosting', 'KNN', 'StackingRegressor']
n_jobs = multiprocessing.cpu_count() // 2
results_ed = {
        'model' : [],
        'value' : [],
        'r2':[],
        'time' : []
    }
for i in tqdm(range(10)):
      df_train = datascaled.sample(frac = 0.7)
      x_train = df_train[['Year','CO2',	'CH4', 'N2O',	'HFCs',	'PFCs',	'NF3',	'SF6']]
      y_train = df_train[['EMISIONES']]

      df_test = datascaled.drop(df_train.index)
      x_test = df_test[['Year','CO2',	'CH4', 'N2O',	'HFCs',	'PFCs',	'NF3',	'SF6']]
      y_test = df_test[['EMISIONES']]
      for m in models_ed:
          start = timer()
          model = fit(x_train, y_train.values.ravel(), m, n_jobs = n_jobs)
          end = timer()
          y_pred = model.predict(x_test)
          results_ed['model'].append(m)
          results_ed['value'].append(RMSE(y_test, y_pred, squared = False))
          results_ed['r2'].append(r2(y_test, y_pred))
          results_ed['time'].append(end - start)


df_results_ed = pd.DataFrame(results_ed)
df_results_ed = df_results_ed.sort_values('value')
```

## ARIMA

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.arima_model import ARMAResults
from statsmodels.tsa.stattools import kpss
from statsmodels.tsa.stattools import adfuller
```

```python
timeseries = data3["EMISIONES"]
timeseries
```

```python
# Esta función revisa si la series es estacionaria, si no lo es diferencía para volverla estacionaria. Devuelve el numero de diferenciaciones hechas.
def stationary(timeseries):
    numediff = 0
    stationary = False
    while stationary == False:
      adf_test = adfuller(timeseries, autolag = 'AIC')
      kpss_test = kpss(timeseries)
      if (adf_test[1] < 0.05 and kpss_test[1] > 0.05) == True:
        stationary = True
      else:
        timeseries = (timeseries - timeseries.shift(1)).dropna()  #diferenciamos para hacer estacionaria
        numediff = numediff+1
    return numediff
```

```python
d = stationary(timeseries)
d
```

```python
aux = 1000000
p = 0
q = 0
for i in tqdm(range(7)):
  for j in (range(7)):
    if (j+i==0) : continue
    modelo_arima = ARIMA(timeseries, order= (i,d,j))
    modelo_fit = modelo_arima.fit()
    if modelo_fit.aic < aux:
      aux = modelo_fit.aic
      p = i
      q = j
```

```python
modelo_arima = ARIMA(timeseries, order= (p,d,q))
modelo_fit = modelo_arima.fit()
modelo_fit.predict()

plt.figure()
plt.plot(data3.Year,timeseries)
plt.plot(data3.Year,modelo_fit.predict())

plt.show()
```
