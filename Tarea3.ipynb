## Calculo de medidas de tendencia central y disperción con solo Python

### Datos auxiliares

Para estos ejercicios se usarán los valores de la columna 6 (CO2) pero removiendo los datos 'nan':

```python
aux = data.iloc[:,6].copy()
aux = aux.dropna()
aux = list(aux)
```

### Media

Podemos calcular  la media con la suma de los datos dividido entre el tamaño de la columna:

```python
media = sum(aux) / len(aux)

media
```

### Mediana

Para la mediana primero se deben ordenar los datos luego se toma el dato que se encuentre a la mitad. Si es un número par se toma el promedio de los dos centrales:

```python
aux_sorted = sorted(aux)

if len(aux)%2 == 0:
    mediana = (aux_sorted[int(len(aux)/2)] + aux_sorted[int((len(aux)/2)+1)])/2
if len(aux)%2 == 1:
    mediana = aux_sorted[int((len(aux)+1)/2)]

mediana
```

Se agregó la función *int* dentro de los indices ya que a oesar de que los valores son en teoría enteros la variable en Python era flotante por lo que se debe convertir para que funcione el código

### Moda

Obtenemos primero una lista de frecuencias

```python
frec = []
for i in range(0,len(aux)):
    n=0
    for j in range(0,len(aux)):
        if aux[i] == aux[j]:
            n=n+1
    
    frec.append(n)

frec
```

Luego obtenemos el número que tenga mas repeticiones para obtener la moda:

```python
for i in range(0, len(aux)):
    if frec[i] == max(frec):
        n = aux[i]

n
```

### Varianza

Se usa la formula de la varianza:

```python
varianza = 0

for i in range(0,len(aux)):
    varianza = varianza + ((aux[i]-media)**2)

varianza = varianza/(len(aux)-1)

varianza
```

### Desviación estándar

Por definicion es la raiz cuadrada de la varianza:

```python
s = varianza ** (1/2)
s
```

### Sesgo

Usamos la formula:

```python
sesgo = 0

for i in range(0,len(aux)):
    sesgo = sesgo + (((aux[i]-media)**3)*len(aux))

sesgo = sesgo/((len(aux)-1)*(len(aux)-2)*(s**3))

sesgo
```

Se ordenan los datos, se divide la cantidad de elementos entre la cantidad de cuantiles que se desean y se imprimen los elementos correspondientes:

```python
# cuartiles
cuartiles = []
for i in range(1,4):
    if len(aux)%2 == 1:
        q = aux_sorted[int((((len(aux)+1)/4)*i)-1)] # n impar
        cuartiles.append(q)
    if len(aux)%2 == 0:
        q = aux_sorted[int((((len(aux))/4)*i)-1)] # n par
        cuartiles.append(q)

cuartiles.append(aux_sorted[len(aux)-1])
cuartiles
```

## Datos paramétricos (Pruebas de normalidad)

Para saber si los datos son o no paramétricos, revisamos su normalidad.

Primero importamos las librerias a utilizar:

```python
import numpy as np
from scipy.stats import shapiro
```

Revisamos cada una de las columnas que sabemos son numéricas las cuales son de la columna 6 a la 35.

Usamos la función *shapiro* para obtener el p valor de la prueba y comparar con un alfa del 0.05.

**NOTA:** Al realizar la prueba se descubrió que la columna 34 está vacía por lo que se elimina de los datos.

```python
data.pop("EMISIONES(sin 3B y 3D)(Gg en CO2e)")
data
```

```python
normal = []
nonormal=[]

for i in range(6,35):
    if shapiro(data.iloc[:,i].dropna()).pvalue <= 0.05:
        normal.append(i)
    else:
        nonormal.append(i)

print("Normales:", normal)
print("No normales:", nonormal)
```
