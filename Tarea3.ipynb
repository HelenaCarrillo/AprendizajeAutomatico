## Calculo de medidas de tendencia central y disperción con solo Python

### Datos auxiliares

Para estos ejercicios se usarán los valores de la columna 6 (CO2) pero removiendo los datos 'nan':

```python
aux = data.iloc[:,6].copy()
aux = aux.dropna()
aux = list(aux)
```

### Media

Podemos calcular  la media con la suma de los datos dividido entre el tamaño de la columna:

```python
media = sum(aux) / len(aux)

media
```

### Mediana

Para la mediana primero se deben ordenar los datos luego se toma el dato que se encuentre a la mitad. Si es un número par se toma el promedio de los dos centrales:

```python
aux_sorted = sorted(aux)

if len(aux)%2 == 0:
    mediana = (aux_sorted[int(len(aux)/2)] + aux_sorted[int((len(aux)/2)+1)])/2
if len(aux)%2 == 1:
    mediana = aux_sorted[int((len(aux)+1)/2)]

mediana
```

Se agregó la función *int* dentro de los indices ya que a oesar de que los valores son en teoría enteros la variable en Python era flotante por lo que se debe convertir para que funcione el código

### Moda

Obtenemos primero una lista de frecuencias

```python
frec = []
for i in range(0,len(aux)):
    n=0
    for j in range(0,len(aux)):
        if aux[i] == aux[j]:
            n=n+1
    
    frec.append(n)

frec
```

Luego obtenemos el número que tenga mas repeticiones para obtener la moda:

```python
for i in range(0, len(aux)):
    if frec[i] == max(frec):
        n = aux[i]

n
```

### Varianza

Se usa la formula de la varianza:

```python
varianza = 0

for i in range(0,len(aux)):
    varianza = varianza + ((aux[i]-media)**2)

varianza = varianza/(len(aux)-1)

varianza
```

### Desviación estándar

Por definicion es la raiz cuadrada de la varianza:

```python
varianza ** (1/2)
```
